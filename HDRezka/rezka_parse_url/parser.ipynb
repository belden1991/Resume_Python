```python
def parse_page_number(url = "http://rezkly.com/"):
    #url = "http://rezkly.com/"
    resp = req.get(url)
    soup = BeautifulSoup(resp.text, 'lxml')
    list_item = soup.find_all("div", {"class": "b-content__inline_item"})
    url_item = soup.find_all("div", {"class": "b-navigation"})
    len_url = url_item[0].find_all("a")[-2].text
    print(len_url)
    return(len_url)

def parse_movies(url = "http://rezkly.com/"):
    #url = "http://rezkly.com/"
    global len_url
    global url_link
    global movie_cats
    global num_series
    global name
    
    resp = req.get(url)
    soup = BeautifulSoup(resp.text, 'lxml')
    list_item = soup.find_all("div", {"class": "b-content__inline_item"})
    name_item = soup.find_all("div", {"class": "b-content__inline_item-link"})
    for i in range(len(list_item)):
        url_link.append(list_item[i].find("a").get("href"))
        movie_cats.append(list_item[i].find("i").text)
        if len(list_item[i].find_all("span")) == 2:
            num_series.append(list_item[i].find_all("span")[1].text)
        else:
            num_series.append("")
        name.append(name_item[i].find("a").text)
        #print(i)
        #print("__"*20)
        #print(name)
```
