## Линейная регрессия

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing

y = df['price']
X = df.drop('price', axis = 1)
X_my = X.iloc[:322]
y_my = y.iloc[:322]


min_max_scaler = preprocessing.MinMaxScaler()
df_scaled = min_max_scaler.fit_transform(X)
df1 = pd.DataFrame(df_scaled)
df1,X
```

```python
print(X.head())

print('_'*30)
print('без нашей машины')
print('_'*30)

reg = LinearRegression().fit(X_my, y_my)
print(reg.score(X_my, y_my))
print(reg.coef_)
print(reg.intercept_)

print(X_my.shape,X.shape)
reg.predict(X)[-1]


print('_'*30)
print('вместе с нашей машиной')
print('_'*30)

reg = LinearRegression().fit(X, y)
print(reg.score(X, y))
print(reg.coef_)
print(reg.intercept_)

print(X_my.shape,X.shape)
reg.predict(X)[-1]
```

## Дерево решений

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(max_depth=10)
regressor.fit(X_my, y_my)
print(regressor.score(X_my,y_my))
for i in range(1,30):
    regressor = DecisionTreeRegressor(max_depth=i)
    regressor.fit(X_my, y_my)
    print(i,regressor.score(X_my,y_my),regressor.predict(X)[-1])
```

На мой взгляд, данных было мало для обучения моделей, универсалов так вообще было около 5% от всех данных, однако даже с этими данными можно было хорошо поработать, и тем самым улучшить точность модели. Например, можно было бы все таки убрать одну из характеристик (объем двигателя или его мощность), можно было бы применить к ним dummy-кодирование, или например округлить мощность до значений которые есть в спецификации (наверняка кто-то указывал мощность от себя, или просто мог ошибиться), ну и дальнейшем применить к ним различные модели обучения. Однако не думаю что на этих данныхможно было бы добиться впечатлюящих результатов.

Я изначально выбрал именно линейную модель как основную, так как посчитал, что ее проще всего будет интерпретировать в моем случае. Можно было применить другие методы и модели для улучшения качества предсказания, однако мне не нужна была высокая точность, так как машину я не продавал, и мне было просто интересно в какую цену ее оценит страховая фирма. На основе выводов моделей, а также своих собственных суждениях, я бы оценил ее на тот момент в 320 тыс рублей. Потому что, помимо сухих цифр, на стоимость влияет также очень много других факторов (техническое состояние, комплетация, внешний вид, город и т.д.).

Что касается моделей которые были использованы:
На мой взгляд линейная модель очень хорошо себя проявила, так же, если бы мне нужна была хорошая точность, я бы присмотрелся к модели CatBoost. Что касается деревьев, модель достаточно точна, и может быть использована, тем более если подобрать параметры, но для меня она была вспомогательной моделью, чтобы оценить и сравнить линейную модель. Так же можно было объединить несколь моделей, и тем самым улучшить точность еще сильнее
